```python
from game.game import PrisonersDilemmaGame
from tournament.tournament import Tournament
from strats.tit_for_tat import TitForTat
from strats.random import RandomStrategy
from strats.friedman import Friedman
from strats.joss import Joss
from strats.harrington import Harrington
from strats.always_cooperate import AlwaysCooperate
from strats.always_defect import AlwaysDefect
```

Games

```python
g1 = PrisonersDilemmaGame(TitForTat(), RandomStrategy(), n=50)
g2 = PrisonersDilemmaGame(Friedman(), Joss(), n=50)
g3 = PrisonersDilemmaGame(Harrington(), TitForTat(), n=50)
g4 = PrisonersDilemmaGame(AlwaysCooperate(), AlwaysDefect(), n=50)
g5 = PrisonersDilemmaGame(AlwaysCooperate(), TitForTat(), n=50)
g6 = PrisonersDilemmaGame(AlwaysCooperate(), RandomStrategy(), n = 50)
```

```python
g1.play_game()
g2.play_game()
g3.play_game()
g4.play_game()
g5.play_game()
g6.play_game()
```

```python
g1.plot_payoffs()
```

```python
g6.plot_payoffs()
```

Tournament

```python
participants = [
    TitForTat(), RandomStrategy(), Friedman(), Joss(),
    Harrington(), AlwaysCooperate(), AlwaysDefect()
]

tournament = Tournament(participants, n_rounds=200)
tournament.run_tournament()
```
